```python
import pandas as pd
```

```python
df = pd.read_csv("/content/games.csv")
```

```python
df
```

```python
df.info()
```

```python
df['release_date']= pd.to_datetime(df['release_date'])#,format="%Y-%m-%d")
```

```python
df['developers']=df['developers'].str.replace(r"[\[\]']", "", regex=True)
df['publishers']=df['publishers'].str.replace(r"[\[\]']", "", regex=True)
df['genres']=df['genres'].str.replace(r"[\[\]']", "", regex=True)
df['supported_languages']=df['supported_languages'].str.replace(r"[\[\]']", "", regex=True)


df
```

```python
df.isna().sum()
df['genres'].fillna("", inplace=True)
df['publishers'].fillna("", inplace=True)
df['developers'].fillna("", inplace=True)

df
```
