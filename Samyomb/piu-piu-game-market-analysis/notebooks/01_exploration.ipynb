```python
import pandas as pd
df = pd.read_csv('/content/ventes_consoles_globales_v2.csv')
df_ca = df.groupby("Console")[['Constructeur','Revenu_brut_estime_MdsUSD']].sum()
```

```python
df.head()
```

```python
import plotly.express as px
fig = px.bar(
    df_ca,
    x="Constructeur",
    y="Revenu_brut_estime_MdsUSD",
    color=df_ca.index,
    barmode="group",
    title="Revenus par constructeur et par console"
)
fig.show()
```

```python
fig2 = px.bar(
    df,
    x="Année_sortie",
    y="Ventes_millions",
    color='Console',
    title="Ventes de consoles dans le temps"
)
fig2.show()
```
